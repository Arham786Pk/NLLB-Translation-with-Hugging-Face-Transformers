```python
from transformers import pipeline
```

```python
pipe = pipeline("translation", model="facebook/nllb-200-1.3B")
```

```python
result = pipe("Hello, how are you?", src_lang="eng_Latn", tgt_lang="urd_Arab")
print(result[0]['translation_text'])
```
